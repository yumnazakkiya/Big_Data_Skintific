```python
import pandas as pd

# Load the uploaded CSV file
df = pd.read_csv("cushion_gold_skintific.csv", delimiter=';', encoding='utf-8-sig')


# Display the first few rows to understand the structure
df.head()
```

```python
# Tampilkan nama kolom dan beberapa isi review sebelum dibersihkan
df.columns, df["Ulasan"].head()
```

```python
def bersihkan_teks(teks):
    if pd.isna(teks):
        return teks
    # Hanya sisakan huruf dan spasi
    return ''.join(char for char in teks if char.isalpha() or char.isspace())
```

```python
df["Ulasan Bersih"] = df["Ulasan"].apply(bersihkan_teks)
```

```python
print(df.head())
```

```python
# Cek jumlah nilai null di setiap kolom
print(df.isnull().sum())
```

```python
df["Ulasan Bersih"] = df["Ulasan"].str.lower()
```

```python
import string
df["Ulasan Bersih"] = df["Ulasan Bersih"].str.translate(str.maketrans('', '', string.punctuation + '0123456789'))
```

```python
import nltk
nltk.download('punkt', force=True)  # paksa download ulang
```

```python
df["Tokenized"] = df["Ulasan Bersih"].apply(
    lambda x: x.lower().split() if isinstance(x, str) else []
)
```

```python
from nltk.corpus import stopwords
nltk.download('stopwords')
stop_words = set(stopwords.words('indonesian'))
df["Filtered"] = df["Tokenized"].apply(lambda tokens: [word for word in tokens if word not in stop_words])
```

```python
from Sastrawi.Stemmer import StemmerFactory

# Membuat objek StemmerFactory
factory = StemmerFactory.StemmerFactory()

# Membuat stemmer dari objek factory
stemmer = factory.create_stemmer()

# Mengaplikasikan stemming pada DataFrame
df["Stemmed"] = df["Filtered"].apply(lambda tokens: [stemmer.stem(word) for word in tokens])
```

```python
df["Final Text"] = df["Stemmed"].apply(lambda tokens: ' '.join(tokens))
```
